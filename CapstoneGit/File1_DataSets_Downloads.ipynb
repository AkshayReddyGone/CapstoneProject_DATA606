```python
import os
import requests
from bs4 import BeautifulSoup

# Set up the base URL and download directory
base_url = 'https://aqs.epa.gov/aqsweb/airdata/download_files.html'
download_dir = '/Users/akshayreddy/Documents/AirQualityData'

# Create download directory if it doesn't exist
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

# Send a GET request to the webpage
response = requests.get(base_url)
response.raise_for_status()

# Parse the webpage content
soup = BeautifulSoup(response.text, 'html.parser')

# File patterns to filter
hourly_data_patterns = ['hourly_']
daily_aqi_patterns = ['daily_aqi_']

# Function to download files based on patterns
def download_files(patterns):
    for pattern in patterns:
        # Find all links that match the pattern
        for link in soup.find_all('a', href=True):
            if pattern in link['href']:  # Check for the specific pattern
                # Extract the year from the filename
                year = link['href'].split('_')[-1].split('.')[0]
                if year.isdigit() and 2015 <= int(year) <= 2024:
                    file_url = f"https://aqs.epa.gov/aqsweb/airdata/{link['href']}"
                    file_name = link['href']
                    download_path = os.path.join(download_dir, file_name)

                    # Download the file
                    print(f"Downloading {file_name}...")
                    file_response = requests.get(file_url, stream=True)
                    file_response.raise_for_status()

                    # Save the file
                    with open(download_path, 'wb') as file:
                        for chunk in file_response.iter_content(chunk_size=8192):
                            file.write(chunk)

                    print(f"Downloaded {file_name} to {download_path}")

# Download hourly data files
download_files(hourly_data_patterns)

# Download daily AQI files
download_files(daily_aqi_patterns)
```

```python
import os
import requests
from bs4 import BeautifulSoup

# Set up the base URL and download directory
base_url = 'https://aqs.epa.gov/aqsweb/airdata/download_files.html'
download_dir = '/Users/akshayreddy/Documents/AirQualityData'

# Create download directory if it doesn't exist
if not os.path.exists(download_dir):
    os.makedirs(download_dir)

# Send a GET request to the webpage
response = requests.get(base_url)
response.raise_for_status()

# Parse the webpage content
soup = BeautifulSoup(response.text, 'html.parser')

# File patterns to filter
hourly_data_patterns = ['hourly_']
daily_aqi_patterns = ['daily_aqi_']

# Function to download files based on patterns
def download_files(patterns):
    for pattern in patterns:
        for link in soup.find_all('a', href=True):
            if pattern in link['href']:
                year = link['href'].split('_')[-1].split('.')[0]
                if year.isdigit() and 2015 <= int(year) <= 2024:
                    file_url = f"https://aqs.epa.gov/aqsweb/airdata/{link['href']}"
                    file_name = link['href']
                    download_path = os.path.join(download_dir, file_name)

                    try:
                        if os.path.exists(download_path):
                            print(f"{file_name} already exists. Skipping...")
                            continue  # Skip if the file already exists

                        print(f"Downloading {file_name}...")
                        file_response = requests.get(file_url, stream=True)
                        file_response.raise_for_status()

                        with open(download_path, 'wb') as file:
                            for chunk in file_response.iter_content(chunk_size=8192):
                                file.write(chunk)

                        print(f"Downloaded {file_name} to {download_path}")
                    except requests.exceptions.Timeout:
                        print(f"Timeout occurred while downloading {file_name}. Skipping...")
                    except Exception as e:
                        print(f"An error occurred: {e}. Skipping...")

# Download hourly data files
download_files(hourly_data_patterns)

# Download daily AQI files
download_files(daily_aqi_patterns)
```
